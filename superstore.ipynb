```python
# Which products categories and sub categories are most profitable?
```

```python
profit_by_category = df.groupby(['Category','Sub-Category'])['Profit'].sum().sort_values(ascending=False)
profit_by_category
```

```python
profit_by_category.plot(kind='bar', figsize = (12,6), title='profit by category and sub-category')
plt.ylabel('Total profit')
plt.tight_layout()
plt.show()
```

```python
# Which regions generate the most sales and profit?
```

```python
region_withmost_sales_profit= df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
region_withmost_sales_profit
```

```python
region_withmost_sales_profit.plot(kind = 'bar', figsize = (12,6), title = 'most sales & profit by region')
plt.ylabel('sales')
plt.tight_layout()
plt.show
```

```python
# Is there a correlation between discount and profit?
```

```python
correlation_discount_profit = df.groupby('Discount')['Profit'].sum().sort_values(ascending=False)
correlation_discount_profit
```

```python
correlation_discount_profit.plot(kind='bar', figsize = (12,6), title='correlation between discount & profit')
plt.ylabel('profit')
plt.tight_layout()
plt.show
```
